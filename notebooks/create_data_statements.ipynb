```python
import os
import pandas as pd
import xlrd  # dong nay de doc file xlsm
import glob
pd.set_option('display.max_columns',500,'float_format', '{:,.2f}'.format) #hien het cac dong, format dau phay, 2 so 0
pd.set_option('display.max_rows', 500)

#path_CDKT= '../data/raw/yearly'# use your path
#CDKT_files = glob.glob(path_CDKT + "/*.xlsm")
#li_CDKT = []
#for filename in CDKT_files:
#    df = pd.read_excel(filename, sheet_name=None, header=None)

path_CDKT= '../data/raw/yearly' # use your path
# data/raw/Bank/yearly/FiinPro_BCTC_MBB_hopnhat_2021_10_tyvnd_08_02_2022144946739.xlsm
b = '../data/raw/Bank/yearly/FiinPro_BCTC_ACB_hopnhat_2021_10_tyvnd_08_02_202214420912.xlsm' #path
CDKT_files = glob.glob(path_CDKT + "/*.xlsx")
df = pd.read_excel(b, sheet_name=None, header=None) # doc het cac sheet
```

```python
df_BS = df['Cân đối kế toán'] # lay sheet CDKT
Ma_CP = df_BS.loc[7,1]   # lay ma cp o dong 8, cot B
Ma_CP
#df_BS.dropna(how='all', inplace = True)
row_tieu_de = int(df_BS[df_BS[0] == 'Tiêu đề'].index[0])# tim row Tiêu đề
begin_BS = row_tieu_de + 6
row_CTCP_Fin_row = int(df_BS[df_BS[0] == 'CÔNG TY CỔ PHẦN FIINGROUP'].index[0])# tim row CÔNG TY CỔ PHẦN FIINGROUP
end_BS = row_CTCP_Fin_row - 1
BS = df_BS.iloc[begin_BS:end_BS]
BS.iloc[0,0]= "items"
BS.iloc[1,0]= "Balance_sheet"
BS.columns = BS.iloc[0] # lay dong dau tien lam ten column
BS.drop(BS.index[0],inplace= True) # bo dong dau tien
BS.set_index('items', inplace = True) # set index cột items
BS_drop = BS.drop('2022F',axis=1) # loại cột Forecást 2022
#BS growth
BS_pct= BS_drop.loc['TỔNG TÀI SẢN':].pct_change(axis = 'columns')*100
#BS_pct
BS_growth = BS_pct.loc[['TỔNG TÀI SẢN','     Cho vay khách hàng','     Tiền gửi của khách hàng','Vốn chủ sở hữu']]
BS_growth
```

```python
BS_CS = BS.loc['TỔNG TÀI SẢN':]*100/BS.loc['TỔNG TÀI SẢN']  # BS common size
BS_CS
```

```python
df_IS = df['Kết quả Kinh doanh'] # Kết quả Kinh doanh
row_tieu_de = int(df_IS[df_IS[0] == 'Tiêu đề'].index[0])# tim row Tiêu đề
begin_IS = row_tieu_de + 6
row_CTCP_Fin_row = int(df_IS[df_IS[0] == 'CÔNG TY CỔ PHẦN FIINGROUP'].index[0])# tim row CÔNG TY CỔ PHẦN FIINGROUP
end_IS = row_CTCP_Fin_row - 1
IS = df_IS.iloc[begin_IS:end_IS]
IS.iloc[0,0]= "items"
IS.iloc[1,0] = "Income_statement"
IS.columns = IS.iloc[0] # lay dong dau tien lam ten column
IS.drop(IS.index[0],inplace= True) # bo dong dau tien
IS.set_index('items', inplace = True) #
#IS
IS_drop = IS.drop('2022F',axis=1) # loại cột Forecást 2022
total_revenue = IS_drop.loc['Thu nhập lãi và các khoản thu nhập tương tự'] + IS_drop.loc['     Thu nhập từ hoạt động dịch vụ'] + IS_drop.loc['Lãi/(lỗ) thuần từ hoạy động kinh doanh ngoại hối và vàng'] + IS_drop.loc['Lãi/(lỗ) thuần từ mua bán chứng khoán kinh doanh'] + IS_drop.loc['     Thu nhập từ hoạt động khác'] + IS_drop.loc['Thu nhập từ góp vốn, mua cổ phần']
#total_revenue
IS_pct= IS_drop.loc['Thu nhập lãi và các khoản thu nhập tương tự':].pct_change(axis = 'columns')*100
#IS_pct
IS_growth = IS_pct.loc[['Thu nhập lãi thuần','     Thu nhập từ hoạt động dịch vụ','Chi phí hoạt động','LN thuần từ hoạt động kinh doanh trước CF dự phòng rủi ro tín dụng','Tổng lợi nhuận trước thuế','Lợi nhuận sau thuế','Lãi cơ bản trên cổ phiếu (VND)']]
#IS_growth
ratio_growth = pd.concat([BS_growth, IS_growth], axis=0)
ratio_growth.rename_axis("ratio growth", inplace=True)
growth =  ratio_growth.T
#growth
idx = 0
growth.insert(loc = idx, column= 'CP', value= Ma_CP)
growth
```

```python
IS_CS = IS_drop.loc['Thu nhập lãi và các khoản thu nhập tương tự':]*100/IS_drop.loc['Tổng thu nhập hoạt động']
IS_CS
```

```python
BS_TS_BQ = BS_drop.loc['TỔNG TÀI SẢN'].rolling(2, axis = 0).mean() # Tổng TS bình quân để tính chỉ số hiệu quả
BS_TSSL= BS_drop.loc[['Tiền gửi tại Ngân hàng nhà nước Việt Nam','     Tiền gửi tại các TCTD khác','     Cho vay khách hàng','     Chứng khoán đầu tư sẵn sàng để bán','     Chứng khoán đầu tư giữ đến ngày đáo hạn']].sum() # TS sinh lãi bình quân
BS_TSSL_BQ = BS_TSSL.rolling(2, axis = 0).mean() # Tài sản sinh lãi bình quân (theo năm)
BS_VCSH_BQ = BS_drop.loc['Vốn chủ sở hữu'].rolling(2, axis = 0).mean() # Tổng TS bình quân để tính chỉ số hiệu quả
HQ_TSBQ = IS_drop.loc[['Tổng thu nhập hoạt động','Chi phí hoạt động']]*100/BS_TS_BQ # chỉ số Hiệu Quả trên TSBQ
HQ_TSBQ.rename_axis("perfomance", inplace = True)
HQ_TSBQ.rename(index= {'Tổng thu nhập hoạt động':'Tổng TN HĐ/Tổng TS BQ (%)', 'Chi phí hoạt động':'Chi phí hoạt động/Tổng TS BQ(%)'}, inplace = True)
HQ_TSBQ
```

```python
df_CF = df['Lưu chuyển tiền tệ'] # Lưu chuyển tiền tệ
row_tieu_de = int(df_CF[df_CF[0] == 'Tiêu đề'].index[0]) #tim tieu de
begin_CF = row_tieu_de + 6
row_gian_tiep = int(df_CF[df_CF[0] == 'Gián Tiếp '].index[0])# tim
row_CTCP_Fin_row = int(df_CF[df_CF[0] == 'CÔNG TY CỔ PHẦN FIINGROUP'].index[0])# tim row CÔNG TY CỔ PHẦN FIINGROUP
end_CF = row_CTCP_Fin_row - 1
CF_Dir = df_CF.iloc[begin_CF:row_gian_tiep-1] # Cashflow statement truc tiep
CF_Dir.iloc[0,0]= "items"
CF_Dir.iloc[1,0]= "CF trực tiếp"
CF_Dir.columns = CF_Dir.iloc[0] # lay dong dau tien lam ten column
CF_Dir.drop(CF_Dir.index[0],inplace= True) # bo dong dau tien
CF_Dir.set_index('items', inplace = True) #
CF_Dir
```

```python
CF_inDir = df_CF.iloc[row_gian_tiep:end_CF] # Cashflow statement gian tiep
CF_inDir.iloc[0,0]= "items"
CF_inDir.iloc[1,0]= "CF gián tiếp"
CF_inDir.columns = CF_inDir.iloc[0] # lay dong dau tien lam ten column
CF_inDir.drop(CF_inDir.index[0],inplace= True) # bo dong dau tien
CF_inDir.set_index('items', inplace = True) #
CF_inDir
```

```python
df_note = df['Thuyết minh'] # Thuyết minh
row_tieu_de = int(df_note[df_note[0] == 'Tiêu đề'].index[0]) #tim tieu de
begin_note = row_tieu_de + 6
row_CTCP_Fin_row = int(df_note[df_note[0] == 'CÔNG TY CỔ PHẦN FIINGROUP'].index[0])# tim row CÔNG TY CỔ PHẦN FIINGROUP
end_note = row_CTCP_Fin_row - 1
notes = df_note.iloc[begin_note:end_note] # C
notes.iloc[1,0]= "Notes"
notes.iloc[0,0]= "items"
notes.iloc[1,0]= "notes"
notes.columns = notes.iloc[0] # lay dong dau tien lam ten column
notes.drop(notes.index[0],inplace= True) # bo dong dau tien
notes.set_index('items', inplace = True) #
notes.loc['CAR']
```

```python
HQ_TSBQ.loc['TS SinhLai BQ/TSBQ'] =  BS_TSSL_BQ*100/BS_TS_BQ  # TS SinhLai BQ/TSBQ
HQ_TSBQ.loc['Yield on Earning Asset - YOEA'] = IS_drop.loc['Thu nhập lãi và các khoản thu nhập tương tự']*100/BS_TSSL_BQ # Yield on Earning Asset - tỷ suất sinh lời trên các tài sản có sinh lãi
HQ_TSBQ.loc['Chi phí Lãi'] = IS_drop.loc['Chi phí lãi và các chi phí tương tự']*100/BS_TSSL_BQ
HQ_TSBQ.loc['NIM'] = IS_drop.loc['Thu nhập lãi thuần']*100/BS_TSSL_BQ
HQ_TSBQ.loc['chi phí trích lập DP/TSSLBQ'] = IS_drop.loc['Chi phí dự phòng rủi ro tín dụng']*100/BS_TSSL_BQ
HQ_TSBQ.loc['LN thuần/TS SinhLai BQ'] =  IS_drop.loc['Lợi nhuận sau thuế']*100/BS_TSSL_BQ
HQ_TSBQ.loc['ROAA'] =  IS_drop.loc['Lợi nhuận sau thuế']*100/BS_TS_BQ
HQ_TSBQ.loc['ROAE'] =  IS_drop.loc['Lợi nhuận sau thuế']*100/BS_VCSH_BQ
HQ_TSBQ.loc['Leverage_BQ'] =  BS_TS_BQ/BS_VCSH_BQ
HQ_TSBQ.loc['Thu nhập lãi thuần/Tổng TN Hoạt động (%)'] = IS_CS.loc['Thu nhập lãi thuần']
HQ_TSBQ.loc['Lãi/Lỗ thuần từ hoạt động dịch vụ/Tổng TN Hoạt động (%)'] = IS_CS.loc['Lãi/Lỗ thuần từ hoạt động dịch vụ']
HQ_TSBQ.loc['Lãi/(lỗ) thuần từ hoạt động khác/Tổng TN Hoạt động (%)'] = IS_CS.loc['Lãi/(lỗ) thuần từ hoạt động khác']
HQ_TSBQ.loc['CIR - Chi phí hd/Tổng TN hd (%)'] = IS_CS.loc['Chi phí hoạt động']
HQ_TSBQ.loc['LN thuần/Tổng TN Hoạt động (%)'] = IS_CS.loc['Lợi nhuận sau thuế']
HQ_TSBQ.loc['CAR (%)'] = notes.loc['CAR'].astype(float)*100  # chuyen string sang float, do trong note, dong CAR ở dạng string
HQ_TSBQ.loc['VCSH/TS end (%)'] = BS_drop.loc['Vốn chủ sở hữu']*100/BS_drop.loc['TỔNG TÀI SẢN']
HQ_TSBQ.loc['NPL (%)'] = notes.loc[['     Nợ xấu có khả năng mất vốn','     Nợ nghi ngờ','     Nợ dưới tiêu chuẩn']].sum()*100/notes.loc['Các khoản cho vay phân loại theo chất lượng nợ cho vay']
HQ_TSBQ.loc['LLR (%)'] = BS_drop.loc['     Dự phòng rủi ro cho vay khách hàng']*(-100)/notes.loc[['     Nợ xấu có khả năng mất vốn','     Nợ nghi ngờ','     Nợ dưới tiêu chuẩn']].sum()
HQ_TSBQ.loc['LDR (%)'] = BS_drop.loc['Cho vay khách hàng']*100/BS_drop.loc[['     Tiền gửi của khách hàng','     Phát hành giấy tờ có giá']].sum()
HQ_TSBQ.loc['CASA (%)'] = notes.loc[['     Tiền gửi không kỳ hạn','     Tiền gửi ký quỹ']].sum()*100/notes.loc['Các khoản tiền gửi phân loại theo loại tiền gửi']
HQ_TSBQ.loc['BVPS'] = BS_drop.loc['Vốn chủ sở hữu']*10/BS_drop.loc['          Vốn điều lệ']
HQ_TSBQ
```
